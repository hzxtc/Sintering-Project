```python
import numpy as np

for index in range(9,1001):
    with open('DATA', 'a') as f2:
        newEnergy = (-5.187 * index - 1963) / 27.2114 - (-73.66723149)
        newRadius = 0
        if (index >=9 and index <=12 ):
            # using straight line interpolation between Pt 7 in original DATA and Pt 13 in the new given equation
            newRadius = 0.13361621617690464 * index + 3.3413605867616676
        else:
            # Patricia's calculation
            newRadius = 6.2221 * np.log(index) - 10.881
        f2.write('%3i %16.8f %16.8f %16.8f %16.8f \n' %
                    ( index ,  newRadius, 0, 0, newEnergy))
        
```
