```python
import numpy as np
import matplotlib.pyplot as plt
import statistics as st
import os
import csv

steplines = []
with open("metropolis","rt") as file:
    for line in file:
        if line.startswith("s"):
            steplines.append(line.rstrip('\n'))

numberOfCluster = []
steps = []
for i in range(len(steplines)):
    steps.append(i + 1)
    
for j in range(len(steplines)):
    numberOfCluster.append(float(steplines[j].split()[5]))

plt.figure(figsize=(15,8))
plt.plot(steps,numberOfCluster,label="clusters random n =3")
plt.title("50 clusters size 700 K all n =3")
plt.xlabel("step number")
plt.ylabel("number of clusters")
plt.legend()
plt.show()
```
