```python
import numpy as np
import param, math, copy
import matplotlib as mpl
import matplotlib.pyplot as plt

def distance(x1, y1, x2, y2):
    d = ((x1 - x2)**2.0 + (y1 - y2)**2.0)**(0.5)
    return d

Clusters = []
# create a new INIT      
with open('INIT', 'w') as f2:
    f2.write('%3s  %10s  %16s  %15s  %14s\n' % ('Pt', 'R', 'X', 'Y', 'E'))
    
with open('DATA','r') as f4:
    f4.readline()
    data = f4.readlines()
    for line in data:
        Clusters.append([ int(line.split()[0]),
                         float(line.split()[1]),
                         float(line.split()[2]),
                         float(line.split()[3]),
                         float(line.split()[4]) ])
INIT_data = []

currentCLusterSize = 7 # need to change the number for every case

with open('INIT_GM_Pt7_5','r') as f1: # read data from existing file
    f1.readline()
    data = f1.readlines()
    for line in data:
        INIT_data.append([ int(line.split()[0]),
                           float(line.split()[1]),
                           float(line.split()[2]),
                           float(line.split()[3]),
                           float(line.split()[4]) ])
        
Old_data  = copy.deepcopy(INIT_data)


count = 0

while (count < 100):
    if (currentCLusterSize == 2):
        temp = np.random.randint(0, 8)
    elif (currentCLusterSize == 3):
        temp = np.random.randint(8, 20)
    elif (currentCLusterSize == 4):
        temp = np.random.randint(20, 26)
    elif (currentCLusterSize == 5):
        temp = np.random.randint(26, 45)
    elif (currentCLusterSize == 6):
        temp = np.random.randint(45, 49)
    elif (currentCLusterSize == 7):
        temp = np.random.randint(49, 60)
    elif (currentCLusterSize == 8):
        temp = np.random.randint(60, 63)
    elif (currentCLusterSize > 8):
        temp = np.random.randint(0, param.largest_cluster + 55)
    
    
    Enew = Clusters[temp][4]
    
    Rnew = Clusters[temp][1]
    
    with open('INIT', 'a') as f5:
            f5.write('%3i %16.8f %16.8f %16.8f %16.8f \n' %
                    (currentCLusterSize, Rnew, Old_data[count][2], Old_data[count][3], Enew))
    count = count + 1
    
```
